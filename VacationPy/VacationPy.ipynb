```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
```

```python
#Load Data from excercise 1 into DF
cities= "../WeatherPy/output_data/Cities_raw_data.csv"
cities_df = pd.read_csv(cities)
cities_df
```

```python
#configure gmaps and Use the Lat and Lng as locations and Humidity as the weight.Add Heatmap layer to map.
gmaps.configure(api_key=g_key)

cities_df= cities_df.dropna()
locations = cities_df[["Lat", "Lng"]].astype(float)
humidity= cities_df["Humidity"].astype(float)
```

```python
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=humidity)

fig.add_layer(heat_layer)
fig
```

```python
#Narrow down the cities to fit conditions of temps greater than 80, Southern hemisphere between -10 and the equator. 
# Drop any rows will null values
Cities_Vacation_df =  cities_df.loc[(cities_df['Max Temp']>80) & (cities_df['Lat']<0) & (cities_df['Lat']>-10)]
Cities_Vacation_df
```

```python
# Add column for Hotel name and initalize it

Cities_Vacation_df['Hotel Name'] = ""
Cities_Vacation_df
```

```python
# Retrieve the hotel data
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
params = {
    "radius": "5000",
    "type": "lodging",
    "key": g_key,
     
}
# use iterrows to iterate through pandas dataframe
for index, row in Cities_Vacation_df.iterrows():
    
    lat = row["Lat"]
    lng = row["Lng"]
    params["location"] = f"{lat},{lng}"
    city = row["City"]

    # assemble url and make API request
    print(f"Retrieving Results for {city}.")
    response = requests.get(base_url, params=params).json()
    
    # extract results
    results = response['results']
    
    try:
        print(f"Closest hotel to {city} is {results[0]['name']}.")
        Cities_Vacation_df.loc[index, 'Hotel Name'] = results[0]['name']

        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
    print("------------")
```

```python
# Check the outpout 
Cities_Vacation_df
```

```python
# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in Cities_Vacation_df.iterrows()]
locations = Cities_Vacation_df[["Lat", "Lng"]]
```

```python
# map the layers together
hotel_layer = gmaps.symbol_layer(
    locations, fill_color='rgba(0, 150, 0, 0.4)',
    stroke_color='rgba(0, 0, 150, 0.4)', scale=2,
    info_box_content=hotel_info
)


fig = gmaps.figure()

fig.add_layer(heat_layer)
fig.add_layer(hotel_layer)

fig
```
